```python
#Entrenamiento ML sencillo
#A partir de las diferentes características del vino intentar predecir su calidad

import pandas as pd 
import mlflow
import mlflow.sklearn

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

#Cargo los datos del archivo
wine_samples = pd.read_csv('data/wine_samples.csv') 
```

```python
#Separar características y objetivo
X = wine_samples.drop('quality', axis=1)
y = wine_samples['quality']


#Dividir datos entre entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=99)

#Configurar MLflow
mlflow.set_experiment("Wine Quality Prediction")
mlflow.autolog()


# Iniciar un experimento MLflow
with mlflow.start_run():

    # Crear y entrenar el modelo con los datos de entrenamiento
    linear_regression_model = LinearRegression()
    linear_regression_model.fit(X_train, y_train)
    
    # Hacer predicciones
    y_pred = linear_regression_model.predict(X_test)
    
    # Calcular métricas
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    linear_regression_model.score(X_test,y_test)
    
    
```
